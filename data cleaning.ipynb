# To clean unwanted data from dataset

# we use striping from left to right

```python
df['Last_Name']=df['Last_Name'].str.lstrip("...")
df
```

# we use striping from right to left

```python
df['Last_Name']=df['Last_Name'].str.rstrip("...")
```

```python
df
```

```python
df['Phone_Number']=df['Phone_Number'].str.replace('[^a-zA-Z0-9]','')
df
```

```python
df['Phone_Number'] = df['Phone_Number'].fillna('').apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```
